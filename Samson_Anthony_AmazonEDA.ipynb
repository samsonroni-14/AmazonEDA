## 3. ***Data Wrangling***

### Data Wrangling Code

````python
#copy of original dataset
titlesc=data1.copy()
creditsc=data2.copy()

#drop rows with missing title values
titlesc.dropna(subset=['title'], inplace=True)

#Fill numll values in seasons with 0
titlesc['seasons'].fillna(0, inplace=True)


#fill missing IMDb scores and runtimes with the mean
titlesc['imdb_score'].fillna(titlesc['imdb_score'].mean(), inplace=True)
titlesc['runtime'].fillna(titlesc['runtime'].mean(), inplace=True)

#replace missing age certifications with "Unknown"
titlesc['age_certification'].fillna('Unknown', inplace=True)

#replace missing genres with 'Unknown'
titlesc['genres'].fillna('Unknown', inplace=True)

#drop any unnecessary columns if you find any (example below)
titlesc.drop(columns=['description'], inplace=True)

#reset Index after cleaning
titlesc.reset_index(drop=True, inplace=True)

#Fill missing IMDB id with a placeholder
titlesc['imdb_id'].fillna('Unknown', inplace=True)

#Filling missing IMDB votes
titlesc['imdb_votes'].fillna(0, inplace=True)

#Replace null values in tmdb_popularity column with mean
titlesc['tmdb_popularity'].fillna(round(titlesc['tmdb_popularity'].mean(),2), inplace=True)

#Replace null values in tmdb_score column with mean
titlesc['tmdb_score'].fillna(round(titlesc['tmdb_score'].mean(),1), inplace=True)

#confirm its clean
print("Null values after wrangling:\n", titlesc.isnull().sum())
````
